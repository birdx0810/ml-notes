```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Basic variables
n = 10000
loc = 0
scale = 0.3
eps = 1e-1
```

## Laplace Distribution

```python
x1 = np.random.laplace(loc=loc, scale=scale, size=n)
x2 = np.random.laplace(loc=loc+1, scale=scale, size=n)
```

```python
plt.hist(x1, bins=100)
plt.hist(x2, bins=100)
plt.show()
```

```python
x1 = np.random.normal(loc=loc, scale=scale, size=n)
x2 = np.random.normal(loc=loc+1, scale=scale, size=n)
```

```python
plt.hist(x1, bins=100)
plt.hist(x2, bins=100)
plt.show()
```
